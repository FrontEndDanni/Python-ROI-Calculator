```python
class ROI_calculator():
    
    #set to 0 as it starts at 0, until user adds their values 
    
    def __init__(self):
        self.total_income = 0
        self.total_expenses = 0
        self.cash_flow = 0

    def Income(self):
        
        #ask user for total income
        #add it together
        #slap that bad boy in a while loop
        
        while True:
            user_string = input("Enter your total income or input each source of income, without commas. Type 'done' when finished.").lower().strip()
            if user_string == "done":
                return
            try:
                user_income = int(user_string)
                self.total_income += user_income
            except ValueError:
                print("Please enter valid input.")
                        
    def Expenses(self):
        
        #Ask user if they're paying in cash or taking a mortgage
        #Ask for total amount of expenses
        #Also put in a while loop
        
        print("Enter the total amount of expenses you are working with.")
        while True:
            user_string = input("Please enter your expenses, either one expense at a time or the total. Type 'done' when finished.").lower().strip()
            if user_string == "done":
                return
            
            #try & except: to navigate around ValueError with integer 
            
            try:
                user_expenses = int(user_string)
                self.total_expenses += user_expenses
            except ValueError:
                print("Please enter valid input.")
        
    def Cash_Flow(self):
        
        #Cash flow = income - expenses
        #if expenses exceed income tell user
        
        while True:
            self.cash_flow = self.total_income - self.total_expenses
            if self.cash_flow < 0:
                print(f"Your expenses exceed your income! {self.cash_flow} is your current expense. Enter 'done' if you'd still like to continue. Enter 'income' or 'expenses' to return to either of those steps. Or 'reset' to start over.") 
            else:
                print(f"Your cash flow totals to {self.cash_flow}. If you'd like to continue, enter 'done'. Or else enter 'income' or 'expenses' to head back to either steps to adjust them. If you'd like to reset both, enter 'reset'.")
            flow_choice = input("Enter your choice: ").lower().strip()
            if flow_choice == "income":
                self.Income()
            elif flow_choice == "expenses":
                self.Expenses()
            elif flow_choice == "reset":
                self.total_income = 0
                self.total_expenses = 0
                self.Income()
                self.Expenses()
            elif flow_choice == "done":
                return
            else:
                print("Invalid input, please try again.")
            
    def RoI(self):
        
        # Get total investment amount
        # Total monthly cashflow x 12 = annual cash flow
        # Annual cashflow/totalinvestment = ROI
        
        user_string = input("Please enter your total investment costs including down payment, closing costs, etc. ")
        if user_string == "done":
            return
        try:
            investment_amount = int(user_string)
            self.total_expenses += investment_amount
        except ValueError:
            print("Please enter a valid input.")
        annual_cashflow = self.cash_flow * 12
        RoI = annual_cashflow / investment_amount * 100
        print(f"{RoI}%")
        return RoI
    
#call the functions

    def run(self):
        self.Income()
        self.Expenses()
        self.Cash_Flow()
        self.RoI()

#run it! 

cal = ROI_calculator()
cal.run()
```
